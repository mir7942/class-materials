#도식 최적화(Graphical Solution)

파이썬에서 복잡한 수식을 계산하기 위해서는 [NumPy](https://numpy.org/) 라이브러리를 불러와야 합니다. 그리고 그래프를 그리기 위해서는 [Matplotlib](https://matplotlib.org/) 라이브러리를 불러와야 합니다.

외부 라이브러리를 불러올 때는 *import* 또는 *from...import* 구문을 사용합니다.

```python
import numpy as np # NumPy를 불러오고, 이름을 np로 지정합니다.
import matplotlib.pyplot as plt # Matplotlib의 pyplot를 불러오고, 이름을 plt로 지정합니다.
import matplotlib.patheffects as pe # Matplotlib의 patheffects를 불러옵니다. 이때는 patheffects 이름을 pe로 지정합니다.
```

## 격자 설정

그래프를 그리기 전에 함수의 값을 계산하기 위한 격자 공간을 정의합니다. 이때, NumPy의 [meshgrid](https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html) 함수를 사용합니다.

```python
xvec = np.arange(-1.0, 25.0, 0.5) # -1에서 25까지 0.5간격
yvec = np.arange(-1.0, 25.0, 0.5) # -1에서 25까지 0.5간격
x1, x2 = np.meshgrid(xvec, yvec) # xvec를 x-축, yvec를 y-축으로 하는 격자를 만들고, 각 격자 위치를 x1과 x2에 저장합니다.

print(x1)
print(x2)
```

## 함수값 계산

격자 공간 전체에 대해서 목적함수와 제약조건 함수의 값을 계산합니다.

```python
# x1과 x2에 대해 함수 값 계산
f = -400 * x1 - 600 * x2
g1 = x1 + x2 - 16
g2 = x1 / 28 + x2 / 14 - 1
g3 = x1 / 14 + x2 / 24 - 1
g4 = -x1
g5 = -x2
```

## 그래프 생성

Matplotlib를 이용해 그래프를 그립니다.

Matplotlib에서 새 그래프를 만들기 위해서는 [plt.subplots](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) 함수를 호출합니다.

각 함수의 값이 **0**이 되는 지점을 그리기 위해서 [contour](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html) 함수를 호출합니다.

```python
fig, ax = plt.subplots() # 새 그래프 생성

cg1 = ax.contour(x1, x2, g1, levels=[0], colors='sandybrown', ) # 함수 g1(x1, x2)=0인 지점을 그립니다. 이때, colors의 값으로 색상을 지정합니다.
cg1.set(path_effects=[pe.withTickedStroke(angle=135)]) # 빗금을 쳐서 경계가 잘 보이게 합니다. 이때, angle 값으로 빗금 방향을 변경합니다.

cg2 = ax.contour(x1, x2, g2, [0], colors='orangered') # 함수 g2(x1, x2)=0인 지점을 그립니다. 이때, colors의 값으로 색상을 지정합니다.
cg2.set(path_effects=[pe.withTickedStroke(angle=60)]) # 빗금을 쳐서 경계가 잘 보이게 합니다. 이때, angle 값으로 빗금 방향을 변경합니다.

cg3 = ax.contour(x1, x2, g3, [0], colors='blue') # 함수 g3(x1, x2)=0인 지점을 그립니다. 이때, colors의 값으로 색상을 지정합니다.
cg3.set(path_effects=[pe.withTickedStroke(angle=60)]) # 빗금을 쳐서 경계가 잘 보이게 합니다. 이때, angle 값으로 빗금 방향을 변경합니다.

cg4 = ax.contour(x1, x2, g4, [0], colors='green') # 함수 g4(x1, x2)=0인 지점을 그립니다. 이때, colors의 값으로 색상을 지정합니다.
cg4.set(path_effects=[pe.withTickedStroke(angle=90)]) # 빗금을 쳐서 경계가 잘 보이게 합니다. 이때, angle 값으로 빗금 방향을 변경합니다.

cg5 = ax.contour(x1, x2, g5, [0], colors='cyan') # 함수 g5(x1, x2)=0인 지점을 그립니다. 이때, colors의 값으로 색상을 지정합니다.
cg5.set(path_effects=[pe.withTickedStroke(angle=90)]) # 빗금을 쳐서 경계가 잘 보이게 합니다. 이때, angle 값으로 빗금 방향을 변경합니다.

cf = ax.contour(x1, x2, f, [-8800, -7200, -4800, -2400], colors='black') # 함수 f(x1, x2)가 -8800, -7200, -4800, -2400인 지점을 그립니다.
ax.clabel(cf, fmt="%2.1f", use_clabeltext=True) # 위에서 그린 그래프에 함수 f(x1, x2)의 값을 표시합니다.

ax.set_title('Profit Maximization Problem') # 제목을 표시합니다.
ax.set_xlabel('x1') # x-축 제목을 표시합니다.
ax.set_ylabel('x2') # y-축 제목을 표시합니다.

plt.show() # 그래프를 출력합니다.
```
