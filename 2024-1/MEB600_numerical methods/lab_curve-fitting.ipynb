### ì„ í˜• ìµœì†ŒììŠ¹ë²• êµ¬í˜„

ìš°ë¦¬ëŠ” ë°ì´í„°ë¥¼ ì„ í˜• í•¨ìˆ˜ì— í”¼íŒ…í•  ê²ƒì´ë‹¤. ì„ í˜• í•¨ìˆ˜ë¥¼ ì°¾ëŠ” ë°©ë²•ì€ ë‹¤ìŒê³¼ ê°™ë‹¤.

**ì…ë ¥:** $n$ê°œì˜ ë°ì´í„° $\left(x_1,y_1\right),\left(x_2,y_2\right),\ldots,\left(x_n,y_n\right)$

**ì¶œë ¥:** ì„ í˜• í•¨ìˆ˜ $y=a_0+a_1x$ì˜ ì ˆí¸ $a_0$ì™€ ê¸°ìš¸ê¸° $a_1$

**STEP 1:** ê³„ì‚°ì„ í¸ë¦¬í•˜ê²Œ í•˜ê¸° ìœ„í•´ ì¼ë¶€ ê°’ì„ ë¯¸ë¦¬ ê³„ì‚°í•´ ë†“ëŠ”ë‹¤. ì¦‰,

$$
\sum x_i,\ \sum y_i,\sum x_i^2,\sum{x_iy_i}
$$

ë¥¼ ë¯¸ë¦¬ ê³„ì‚°í•œë‹¤.

**STEP 2:** ë‹¤ìŒ ì‹ì„ ì´ìš©í•´ $a_1$ì„ ê³„ì‚°í•œë‹¤.

$$
a_1=\frac{n\sum{x_iy_i}-\sum x_i\sum y_i}{n\sum x_i^2-\left(\sum x_i\right)^2}\tag{1}
$$

**STEP 3:** $a_1$ê³¼ ë‹¤ìŒ ì‹ì„ ì´ìš©í•´ $a_0$ì„ ê³„ì‚°í•œë‹¤.

$$
a_0=\frac{\sum y_i}{n}-a_1\frac{\sum x_i}{n}\tag{2}
$$

**ì§€ì‹œ: ìœ„ì˜ ì•Œê³ ë¦¬ì¦˜ì„ ì°¸ê³ í•˜ì—¬, ì„ í˜• ìµœì†ŒììŠ¹ë²•ì„ êµ¬í˜„í•˜ëŠ” í•¨ìˆ˜ *linregr*ì„ ì™„ì„±í•˜ì‹œì˜¤.**

```python
def linregr(x, y):
  """
  í•¨ìˆ˜ linregrì€ ì„ í˜• ìµœì†Œì œê³±ì ‘í•© ì•Œê³ ë¦¬ì¦˜ì„ êµ¬í˜„í•œë‹¤.
  ì…ë ¥ ë§¤ê°œë³€ìˆ˜:
    x : ë°ì´í„°ì˜ xê°’ì„ ë‚˜íƒ€ë‚´ëŠ” ë²¡í„°
    y : ë°ì´í„°ì˜ yê°’ì„ ë‚˜íƒ€ë‚´ëŠ” ë²¡í„°
  ì¶œë ¥ ë§¤ê°œë³€ìˆ˜:
    a1 : ì ‘í•©í•œ ì„ í˜• í•¨ìˆ˜ì˜ ê¸°ìš¸ê¸°
    a0 : ì ‘í•©í•œ ì„ í˜• í•¨ìˆ˜ì˜ ì ˆí¸
  """

  # ë°ì´í„°ì˜ ê°œìˆ˜ë¥¼ êµ¬í•œë‹¤.
  n = len(x)

  ## STEP 1: ê³„ì‚°ì„ í¸ë¦¬í•˜ê²Œ í•˜ê¸° ìœ„í•´ ì¼ë¶€ ê°’ì„ ë¯¸ë¦¬ ê³„ì‚°í•´ ë†“ëŠ”ë‹¤.
  sumx = np.sum(x)
  sumy = np.sum(y)
  sumsqx = np.sum(x * x)
  sumxy = np.sum(x * y)

  ## STEP 2: ì„ í˜• í•¨ìˆ˜ì˜ ê¸°ìš¸ê¸° a1ì„ ê³„ì‚°í•œë‹¤.
  #### ì½”ë“œ ì‹œì‘ ####
  ## ë°©ë²•) STEP1ì˜ ê³„ì‚° ê²°ê³¼ë¥¼ ì´ìš©í•´ ì‹ (1)ì˜ ì„ í˜• í•¨ìˆ˜ì˜ ê¸°ìš¸ê¸°ë¥¼ ê³„ì‚°í•œë‹¤.
  a1 = 0
  #### ì½”ë“œ ì¢…ë£Œ ####

  ## STEP 3: ì„ í˜• í•¨ìˆ˜ì˜ ì ˆí¸ a0ì„ ê³„ì‚°í•œë‹¤.
  #### ì½”ë“œ ì‹œì‘ ####
  ## ë°©ë²•) STEP1ê³¼ STEP2ì˜ ê³„ì‚° ê²°ê³¼ë¥¼ ì´ìš©í•´ ì‹ (2)ì˜ ì„ í˜• í•¨ìˆ˜ì˜ ì ˆí¸ì„ ê³„ì‚°í•œë‹¤.
  a0 = 0
  #### ì½”ë“œ ì¢…ë£Œ ####

  return a0, a1
```

ìœ„ì— ë§Œë“  í•¨ìˆ˜ *linregr*ë¥¼ í…ŒìŠ¤íŠ¸ í•œë‹¤. ì•„ë¬´ ë¬¸ì œê°€ ì—†ë‹¤ë©´ ì•„ë˜ ì½”ë“œ ì‹¤í–‰ ì‹œ "**ì„±ê³µ**"ì´ë¼ê³  ì¶œë ¥ë˜ì–´ì•¼ í•œë‹¤.

```python
x = np.array([10, 20, 30, 40, 50, 60, 70, 80])
y = np.array([25, 70, 380, 550, 610, 1220, 830, 1450])
a0, a1 = linregr(x, y)

assert abs(a0 + 234.2857) < 0.0001
assert abs(a1 - 19.4702) < 0.0001

print("ì„±ê³µ!!")
```

### í”¼íŒ… í•¨ìˆ˜ ê³„ì‚°í•˜ê¸°

$$

ë¥¼ ì›ë˜ì˜ ì‹ìœ¼ë¡œ ëŒë ¤ë†“ê² ë‹¤. ì™¼ìª½ì˜ $\log$ë¥¼ ì—†ì• ë©´ ë‹¤ìŒê³¼ ê°™ì´ ì‹ì´ ë³€í™˜ëœë‹¤:

## Lagrange ë³´ê°„ë‹¤í•­ì‹

ë‹¤ìŒ ë°ì´í„°ëŠ” 1920ë…„ì—ì„œ 1990ë…„ê¹Œì§€ì˜ ë¯¸êµ­ ì¸êµ¬ë¥¼ ë°±ë§Œ ëª… ë‹¨ìœ„ë¡œ ë‚˜íƒ€ë‚¸ ê²ƒì´ë‹¤.

|ì—°ë„|1920|1930|1940|1950|1960|1970|1980|1990|2000|
|---|---|---|---|---|---|---|---|---|---|
|ì¸êµ¬(ë°±ë§Œ ëª…)|106.46|123.08|132.12|152.27|180.67|205.05|227.23|249.46|?|

ì´ ë°ì´í„°ë¥¼ ê·¼ê±°ë¡œ 2010ë…„ì˜ ì¸êµ¬ë¥¼ ì˜ˆì¸¡í•´ ë³´ë„ë¡ í•˜ê² ë‹¤. ì°¸ê³ ë¡œ 2010ë…„ì˜ ì‹¤ì œ ì¸êµ¬ëŠ” 281.42ë°±ë§Œ ëª…ì´ì—ˆë‹¤.

### Lagrange ë³´ê°„ë‹¤í•­ì‹ êµ¬í˜„

ì´ë¥¼ Lagrange ë³´ê°„ë‹¤í•­ì‹ì„ ì´ìš©í•´ ì¸êµ¬ë¥¼ ì˜ˆì¸¡í•´ ë³´ê² ë‹¤. ìš°ì„  ì£¼ì–´ì§„ ë°ì´í„°ë¥¼ ë³´ê°„í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì°¾ëŠ” Lagrange ë³´ê°„ë‹¤í•­ì‹ì„ íŒŒì´ì¬ìœ¼ë¡œ ì‘ì„±í•´ ë³´ê² ë‹¤. Lagrange ë³´ê°„ë‹¤í•­ì‹ì„ ì°¾ëŠ” ê³µì‹ì€ ë‹¤ìŒê³¼ ê°™ë‹¤:

$$
f_{n-1}\left(x\right)=\sum_{i=1}^{n}{L_i\left(x\right)f\left(x_i\right)}=L_1\left(x\right)f\left(x_1\right)+\ldots+L_n\left(x\right)f\left(x_n\right) \tag{4}
$$

$$
L_i\ (x)f(x_i\ )=f(x_i\ )\prod_{j=1\\jâ‰ i}^n\frac{(x-x_j)}{(x_i-x_j)}=f(x_i)\times\frac{(x-x_1)}{(x_i-x_1 )}\timesâ€¦\times\frac{(x-x_{i-1})}{(x_i-x_{i-1})}\times\frac{(x-x_{i+1})}{(x_i-x_{i+1})}\timesâ€¦\times\frac{(x-x_n)}{(x_i-x_n )} \tag{5}
$$

ì´ë¥¼ ì•Œê³ ë¦¬ì¦˜ìœ¼ë¡œ ë‚˜íƒ€ë‚´ë©´ ë‹¤ìŒê³¼ ê°™ë‹¤:

**ì…ë ¥:** $n$ê°œì˜ ë°ì´í„° $\left(x_1,y_1\right),\left(x_2,y_2\right),\ldots,\left(x_n,y_n\right)$, ë³´ê°„í•  ìœ„ì¹˜ $ğ‘¥^âˆ—$ì˜ ê°’

**ì¶œë ¥:** $x^\ast$ì—ì„œì˜ $f(x^\ast)$ì˜ ê°’

**STEP 1:** ì‹ (4)ì˜ í•©ê³„ë¥¼ ì €ì¥í•  ë³€ìˆ˜ **s**ë¥¼ 0ìœ¼ë¡œ ì´ˆê¸°í™” ì‹œí‚¨ë‹¤. ì•Œê³ ë¦¬ì¦˜ì´ ëë‚  ë•Œ, **s**ëŠ” $f\left(x^\ast\right)$ì˜ ê°’ì„ ê°–ê²Œ ëœë‹¤.

**STEP 2:** $i=1$ì—ì„œ $i=n$ê¹Œì§€, ì‹ (4)ì„ ì´ìš©í•´ $L_i\left(x^\ast\right)f\left(x_i\right)$ë¥¼ ê³„ì‚°í•˜ì—¬ **s**ì— ë”í•œë‹¤.
*   **STEP 2-1:** ì‹ (5)ì˜ ê³±ì…ˆ ê²°ê³¼ë¥¼ ì €ì¥í•  ë³€ìˆ˜ **product**ë¥¼ $f\left(x_i\right)$ë¡œ ì´ˆê¸°í™” ì‹œí‚¨ë‹¤.
*   **STEP 2-2:** ê° $i$ì— ëŒ€í•´, ì‹ (5)ë¥¼ ì´ìš©í•´
$$
\frac{x-x_j}{x_i-x_j}
$$
ë¥¼ ê³„ì‚°í•˜ì—¬ **product**ì— ê³±í•œë‹¤. ë‹¨, $i\neq j$ì¸ ê²½ìš°ë§Œ ê³„ì‚°í•œë‹¤.

**ì§€ì‹œ: ìœ„ì˜ ì•Œê³ ë¦¬ì¦˜ì„ ì°¸ê³ í•˜ì—¬, Lagrange ë³´ê°„ë‹¤í•­ì‹ì„ êµ¬í˜„í•˜ëŠ” í•¨ìˆ˜ *Lagrange*ë¥¼ ì™„ì„±í•˜ì‹œì˜¤.**

```python
def Lagrange(x, y, xx):
  """
  í•¨ìˆ˜ Lagrangeì€ Lagrange ë³´ê°„ë‹¤í•­ì‹ì„ ì´ìš©í•´ xxì—ì„œì˜ yê°’ì„ ì˜ˆì¸¡í•œë‹¤.
  ì…ë ¥ ë§¤ê°œë³€ìˆ˜:
    x : ë°ì´í„°ì˜ xê°’ì„ ë‚˜íƒ€ë‚´ëŠ” ë²¡í„°
    y : xì˜ ìœ„ì¹˜ì—ì„œì˜ ë°ì´í„°ì˜ yê°’ì„ ë‚˜íƒ€ë‚´ëŠ” ë²¡í„°. ì¦‰ f(x)ì˜ ê°’
  ì¶œë ¥ ë§¤ê°œë³€ìˆ˜:
    yint : xxì—ì„œì˜ ì˜ˆì¸¡í•œ yê°’
  """

  # ë°ì´í„°ì˜ ê°œìˆ˜ë¥¼ êµ¬í•œë‹¤.
  n = len(x)

  if len(y) != n:
    return 'x and y must be of same length'

  ## STEP 1: ì‹ (4)ì˜ í•©ê³„ë¥¼ ì €ì¥í•  ë³€ìˆ˜ së¥¼ 0ìœ¼ë¡œ ì´ˆê¸°í™” ì‹œí‚¨ë‹¤.
  ## ì•Œê³ ë¦¬ì¦˜ì´ ëë‚  ë•Œ, sëŠ” f(xx)ì˜ ê°’ì„ ê°–ê²Œ ëœë‹¤.
  s = 0

  ## STEP 2: i=1ì—ì„œ i=nê¹Œì§€, ì‹ (4)ì„ ì´ìš©í•´ L_i(xx)y(i)ë¥¼ ê³„ì‚°í•˜ì—¬ sì— ë”í•œë‹¤.
  for i in range(n):
    ## STEP 2-1: ì‹ (5)ì˜ ê³±ì…ˆ ê²°ê³¼ë¥¼ ì €ì¥í•  ë³€ìˆ˜ productë¥¼ y(i)ë¡œ ì´ˆê¸°í™” ì‹œí‚¨ë‹¤.
    product = y[i]

    ## STEP 2-2: iì— ëŒ€í•´, ì‹ (5)ë¥¼ ê³„ì‚°í•œë‹¤.ë‹¨, iâ‰ jì¸ ê²½ìš°ë§Œ ê³„ì‚°í•œë‹¤.
    for j in range(n):
      ## STEP 2-2 (ê³„ì†): ì‹ (5)ë¥¼ ì´ìš©í•´ (x-x_j)/(x_i-x_j)ë¥¼ ê³„ì‚°í•˜ì—¬ productì— ê³±í•œë‹¤.
      ## ë‹¨, iâ‰ jì¸ ê²½ìš°ë§Œ ê³„ì‚°í•œë‹¤.
      if i != j:
        #### ì½”ë“œ ì‹œì‘ ####
        ## ë°©ë²•) ì‹ (5)ë¥¼ ì´ìš©í•´ (x-x_j)/(x_i-x_j)ë¥¼ ê³„ì‚°í•˜ì—¬ productì— ê³±í•œë‹¤.
        product = product
        #### ì½”ë“œ ì¢…ë£Œ ####

    s = s + product

  yint = s

  return yint
```

ìœ„ì— ë§Œë“  í•¨ìˆ˜ *Lagrange*ë¥¼ í…ŒìŠ¤íŠ¸ í•œë‹¤. ì•„ë¬´ ë¬¸ì œê°€ ì—†ë‹¤ë©´ ì•„ë˜ ì½”ë“œ ì‹¤í–‰ ì‹œ "**ì„±ê³µ**"ì´ë¼ê³  ì¶œë ¥ë˜ì–´ì•¼ í•œë‹¤.

```python
x = np.array([1, 4, 6, 5])
y = np.log(x)
yy = Lagrange(x, y, 2)
print(yy)

assert abs(yy - 0.6288) < 0.0001

print("ì„±ê³µ!!")
```
