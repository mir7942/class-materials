# 파이썬(Python)이란

1990년 귀도 반 로섬(Guido Van Rossum)이 개발한 인터프리터(interpreter) 언어
* 인터프리터 언어: 실행파일로 변환하지 않고, 한 줄씩 소스코드를 해석해서 실행 결과를 확인할 수 있는 언어

파이썬의 확장자는 .py이나, Colab에서는 .ipynb로 저장된다.

## 특징
* 쉬운 문법
* 무료
* 빠른 개발 속도

## 주요 적용 분야
* 시스템 유틸리티 개발
* GUI 프로그래밍
* 웹 프로그래밍
* 수치 연산
* 데이터베이스 프로그래밍
* 데이터 분석
* 사물 인터넷
* 인공지능 프로그래밍

## if문

```
x = 10
if x == 10: # 줄의 마지막에 콜론(colon)이 붙는다.
  print("10입니다.") # 들여쓰기를 해야 한다.
  print("if 안입니다.")
```

```
x = 10
if x == 10:
print("10입니다.") # 들여쓰기를 하지 않은 경우 오류가 발생한다.
```

파이썬에서의 조건은 다음과 같이 표현한다.
- a **>** b : a가 b보다 크다.
- a **>=** b : a가 b보다 크거나 같다.
- a **<** b : a가 b보다 작다.
- a **<=** b : a가 b보다 작거나 같다.
- a **==** b : a와 b가 같다.
- a **!=** b : a와 b가 다르다.


여러 개의 조건을 표현할 때는 논리 연산자를 사용한다.
- a **and** b : a와 b가 모두 True 일 때, True
- a **or** b : a와 b 중 한 개 이상이 True 일 때, True
- **not** a : a가 True이면 False, a가 False이면 True

```
x = 10
if x == 10:
  print("10입니다.")
else: # 역시 콜론에 주의
  print("10이 아닙니다.")
```

```
x = 10
if x < 10:
  print("10보다 작습니다.")
elif x >= 10 and x < 20:
  print("10이상 20미만입니다.")
elif x >= 20 and x < 30:
  print("20이상 30미만입니다.")
else:
  print("30이상입니다.")
```

```
x = 25
if x > 10:
  if x < 20:
    print("10보다 크고 20보다 작습니다.")
  else:
    print("20보다 큽니다.")
else:
  print("10이하입니다.")
```

## while문

```
i = 0
while i < 10:
  print("Hello World")
  i += 1 # i = i + 1
```

```
i = 0
while True: # 무한 루프
  print(i)
  i += 1
  if i == 10:
    break
```

```
i = 0
while i < 10:
  i += 1

  if i % 2 == 0:
    continue

  print(i)
```

## for문
파이썬의 for문은 다른 프로그래밍 언어와 사용법이 다르므로 **주의**해야 한다.

```
a = [1, 4, 3] # 리스트 a
for i in a: # 리스트 a 안에 있는 원소를 하나씩 가져와 i에 대입
  print(i)
```

```
a = [("김", "병철"), ("홍", "길동"), ("아", "무개")] # 3개의 튜플을 가지는 리스트 a
for i in a: # 이 때, i는 튜플이 된다.
  print(i)
```

```
a = [("김", "병철"), ("홍", "길동"), ("아", "무개")]
for i in a:
  print(i[0] + i[1])
```

```
a = [("김", "병철"), ("홍", "길동"), ("아", "무개")]
for (first, last) in a: # 튜플의 원소를 분리해서 가져올 수 있다.
  print(first+last)
```

```
a = [("김", "병철"), ("홍", "길동"), ("아", "무개")]
for first, last in a: # 튜플의 원소를 분리해서 가져올 때, 괄호를 생략할 수 있다.
  print(first+last)
```

```
dict = {
  "Kim": 179, # 키: "Kim", 값: 179
  "Lee": 168, # 키: "Lee", 값: 168
  "Park": 175 # 키: "Park", 값: 175
}

for key in dict: # 딕셔너리에 대해서는 키를 가져온다.
  print(key)
```

```
for value in dict.values(): # 딕셔너리에 대해 값을 가져오고 싶을 때
  print(value)
```

```
for key, value in dict.items(): # 딕셔너리에 대해 키와 값을 가져오고 싶을 때
  print(key, value)
```

```
for i in range(1, 10): # 1부터 9까지 반복. range()를 사용해야 한다.
  print(i)
```

```
a = ["Kim", "Lee", "Park"]
for index, item in enumerate(a): # 원소의 인덱스와 값을 같이 가져오고 싶을 때
  print(f"index={index}, item={item}")
```

# 함수(function)

```
def add(a, b): # 함수는 def로 선언한다.
  c = a + b
  return c # 함수의 결과

print(add(2, 3))
print(add(a=2, b=3)) # 매개변수를 직접 지정하는 것도 가능
```

```
def add_and_mul(a, b):
  c = a + b
  d = a * b
  return c, d # 튜플을 반환한다. 괄호 사용도 가능하다.

print(add_and_mul(2, 3))

e, f = add_and_mul(2, 3)
print(f"e={e}, f={f}")
```

```
def add(a, b = 0): # 기본값 설정
  c = a + b
  return c

print(add(2))
```

```
def print_world(): # 파라미터가 없는 경우
  print("Hello World")
  # return이 없는 경우

print_world()
```

# 모듈(module)과 패키지(package)
- 모듈: 각종 변수, 함수, 클래스를 담고 있는 .py 파일
- 패키지: 여러 모듈을 묶은 것

## 모듈 불러오기

```
import math # 모듈 math를 불러온다.

print(f"PI={math.pi}") # 모듈 math에 있는 변수 pi를 출력한다.

a = math.sqrt(4.0) # 모듈 math에 있는 sqrt() 함수를 이용한다.
print(f"a={a}")
```

```
import math as m # math를 m으로 줄여 부른다.

print(f"PI={m.pi}") # 모듈 m에 있는 변수 pi를 출력한다.

a = m.sqrt(4.0) # 모듈 m에 있는 sqrt() 함수를 이용한다.
print(f"a={a}")
```

```
from math import pi # 모듈 math에서 변수 pi만 가져온다.

print(f"PI={pi}") # 앞에 math를 붙이지 않아도 된다.
```

```
from math import pi, sqrt # 모듈 math에서 변수 pi와 함수 sqrt만 가져온다.

print(f"PI={pi}") # 앞에 math를 붙이지 않아도 된다.

a = sqrt(4.0) # 앞에 math를 붙이지 않아도 된다.
print(f"a={a}")
```

```
from math import * # 모듈 math에 있는 모든 것을 가져온다.

print(f"PI={pi}") # 앞에 math를 붙이지 않아도 된다.

a = sqrt(4.0) # 앞에 math를 붙이지 않아도 된다.
print(f"a={a}")
```
